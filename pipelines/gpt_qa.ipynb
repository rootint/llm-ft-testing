```python
from openai import OpenAI
from tqdm import tqdm

MAIN_MODEL_NAME = "gpt-4-turbo-2024-04-09"
client = OpenAI()
q_a_pairs = {}
INDEX = 94

for i, essay in enumerate(tqdm(df["content"][INDEX:])):
    response = client.chat.completions.create(
        model=MAIN_MODEL_NAME,
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant. Your mission is to generate insightful question/answer pairs based on the given essay.",
            },
            {
                "role": "user",
                "content": f"The essay: {essay}"
                + """Generate 20 insightful question and answer pairs that encapsulate universal principles, 
                strategic insights, and major ideas relevant to startups and entrepreneurship. 
                These Q&A pairs should be crafted in such a way that they can be understood 
                and informative without access to any specific document or article. 
                The questions should cover broad topics such as growth tactics, funding strategies, 
                challenges and more faced by startups. Each answer must be detailed, clearly explaining the concepts in 
                depth while ensuring relevance to widely observed startup phenomena. Avoid any reference that necessitates having prior 
                access to a particular text. FOR EACH MISTAKE THAT INCLUDES ANY MENTION REQUIRING SPECIFIC COURSE CONTEXT, I WILL KILL ONE KITTEN
                Give the answer in a JSON format - here's an example:
                [{"question": "Why do founders need to engage in activities that don't scale?",
                "answer": "Founders need to engage in unscalable activities to push the initial growth and adoption of their startups, much like manually cranking a car engine before it can run on its own."} ...]
                """,
            },
        ],
        response_format={"type": "json_object"},
    )

    q_a_pairs[i + INDEX] = json.loads(response.choices[0].message.content)
    print(f"Index {i + INDEX} complete! Questions:\n{q_a_pairs[i + INDEX]}")
    with open(f"jsons/{i + INDEX}.json", "w") as f:
        f.write(response.choices[0].message.content)
```

```python
import json
from openai import OpenAI
from tqdm import tqdm
import concurrent.futures
import os

os.environ["https_proxy"] = "socks5://127.0.0.1:1086"

MAIN_MODEL_NAME = "gpt-4-turbo-2024-04-09"
client = OpenAI()
q_a_pairs = {}
# INDEX = 94


def process_essay(name, essay):
    if os.path.exists(f"jsons/{name}.json"):
        print(f"File {name} exists, skipping...")
        return name, None, essay

    response = client.chat.completions.create(
        model=MAIN_MODEL_NAME,
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant. Your mission is to generate insightful question/answer pairs based on the given text.",
            },
            {
                "role": "user",
                "content": f"The essay: {essay}"
                + """Generate 15 insightful question and answer pairs that encapsulate universal principles, 
                strategic insights, and major ideas relevant to startups and entrepreneurship BASED ON THIS ESSAY. 
                These Q&A pairs should be crafted in such a way that they can be understood 
                and informative without access to any specific document or article, but cover its contents.
                Each answer must be detailed, clearly explaining the concepts in 
                depth while ensuring relevance to the given essay. Avoid any reference that necessitates having prior 
                access to a particular text. FOR EACH MISTAKE THAT INCLUDES ANY MENTION REQUIRING SPECIFIC SOURCE CONTEXT, I WILL KILL ONE KITTEN
                If the essay is not related to startups or too political, return an empty json.
                Give the answer in a JSON format - here's an example:
                {"questions_and_answers": [{"question": "What color is the sky?", "answer": "Blue"}, ...]}
                """,
            },
        ],
        response_format={"type": "json_object"},
    )
    # q_a_pairs[essay_index] = json.loads(response.choices[0].message.content)
    with open(f"jsons/{name}.json", "w") as f:
        f.write(response.choices[0].message.content)
    return name, response.choices[0].message.content, essay


essays = list(enumerate(df["content"]))
names = list(df["name"])
with concurrent.futures.ThreadPoolExecutor(max_workers=6) as executor:
    futures = [executor.submit(process_essay, names[i], essay) for i, essay in essays]
    for future in tqdm(concurrent.futures.as_completed(futures), total=len(futures)):
        name, result, essay = future.result()
        print(f"Name {name} complete! Text: {essay}")
```

```python
import pandas as pd
import json
import os

# Path to the directory with JSON files
directory = 'jsons'

# List to store data
data = []

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.json'):  # Check if the file is a JSON file
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r') as file:
            json_data = json.load(file)
            # Extract each pair of questions and answers
            try: 
                for qa in json_data['questions_and_answers']:
                    # Append to the list as a dictionary
                    data.append({
                        'id': filename[:-5],
                        'question': qa['question'],
                        'answer': qa['answer']
                    })
            except Exception as e:
                print(f'{e} in {filename}')

# Create a DataFrame
new_df = pd.DataFrame(data)

# Save to CSV
csv_filename = 'output.csv'
new_df.to_csv(csv_filename, index=False)
print(f'Data saved to {csv_filename}')
```
