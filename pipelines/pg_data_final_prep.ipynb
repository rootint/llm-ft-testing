```python
import pandas as pd

df = pd.read_csv('../data/pg_data_1.csv')

df
```

```python
filtered_df = df[df['content'].str.startswith('.')]

filtered_df
```

```python
def remove_first_three_words(sentence):
    return ' '.join(sentence.split()[1:])

# Apply the function to clean 'content' in the filtered DataFrame
filtered_df['content'] = filtered_df['content'].apply(remove_first_three_words)
filtered_df
```

```python
df.update(filtered_df)

df
```

```python
df['year'] = df['content'].str.extract('(\d{4})')

# Output the DataFrame to see the result
print(df)
```

```python
# Convert 'year' content to integers and intentionally not manage errors, to generate NaNs
df['year'] = pd.to_numeric(df['year'], errors='coerce')  # Converts to float, NaN if conversion fails

# Display rows where 'year' is NaN
nan_rows = df[df['year'].isna()]

print("Rows with NaN in the 'year' field:")
nan_rows
df = df.dropna()

df
```

```python
# Convert 'year' content to integers and intentionally not manage errors, to generate NaNs
df['year'] = pd.to_numeric(df['year'], errors='coerce')  # Converts to float, NaN if conversion fails

# Replace NaN values in 'year' with 0
# df['year'].fillna(0, inplace=True)
df['year'] = df['year'].astype(int)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extracting year as string
# df['year'] = df['content'].str.split().str[1]

# Convert 'year' column to integer
df['year'] = df['year'].astype(int)

# Verify the conversion and data extraction

# Plotting
plt.figure(figsize=(20, 7))
min_year = df['year'].min()
max_year = df['year'].max()
plt.hist(df['year'], bins=range(min_year, max_year + 1), alpha=0.7, edgecolor='black')
plt.title('Frequency of Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.xticks(range(min_year, max_year + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
