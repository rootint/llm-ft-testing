```python
import pandas as pd

df = pd.read_csv('pg_data_1.csv')

df
```

```python
filtered_df = df[df['content'].str.startswith('.')]

filtered_df
```

```python
def remove_first_three_words(sentence):
    return ' '.join(sentence.split()[3:])

# Apply the function to clean 'content' in the filtered DataFrame
filtered_df['content'] = filtered_df['content'].apply(remove_first_three_words)
filtered_df
```

```python
df.update(filtered_df)
```

```python
df = df[['url', 'content', 'type', 'name']]
df['content_len'] = df['content'].str.len()
```

```python
df.to_csv('pg_data_final.csv', index=False)
```
